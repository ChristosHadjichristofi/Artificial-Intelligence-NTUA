### Αλγόριθμος παραγωγής λαβυρίνθων
```
1. Αρχικά όλα τα κελιά περιέχουν εμπόδια και η λίστα των τοίχων είναι κενή.
2. Ένα κελί επιλέγεται ως μέρος του λαβυρίνθου. Οι τοίχοι του κελιού προστίθενται στη λίστα των τοίχων.
3. Όσο η λίστα των τοίχων δεν είναι κενή:
      1. Επιλέγεται τυχαία ένας τοίχος από τη λίστα. Εαν μόνο ένα από τα (δύο) κελιά που χωρίζει ο τοίχος είναι μέρος του λαβυρίνθου, τότε:
            1. Ο τοίχος και το κελί που δεν είναι μέρος του λαβυρίνθου, γίνονται μέρος του λαβυρίνθου.
            2. Οι τοίχοι των νέων κελιών που προστέθηκαν στο λαβύρινθο προστίθενται στη λίστα των τοίχων.
      2. Ο τοίχος αφαιρείται από τη λίστα των τοίχων
```
*ως τοίχοι ενός κελιού θεωρούνται όσα απο τα 4 γειτονικά κελιά (όχι τα διαγώνια) περιέχουν εμπόδιο

````
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter
from IPython.display import HTML

class visualization:
    def __init__(self, S, F):
        '''
          Η μέθοδος αυτή αρχικοποιεί ένα αντικείμενο τύπου visualization.
          Είσοδος: 
          -> S: το σημείο εκκίνσης της αναζήτησης
          -> F: το σημείο τερματισμού
        '''
        self.S = S
        self.F = F
        self.images = []
    
    def draw_step(self, grid, frontier, expanded_nodes):
        '''
          Η συνάρτηση αυτή καλείται για να σχεδιαστεί ένα frame στο animation (πρακτικά έπειτα από την επέκταση κάθε κόμβου)
          Είσοδος: 
          -> grid: Ένα χάρτης τύπου grid
          -> frontier: Μια λίστα με τους κόμβους που ανήκουν στο μέτωπο της αναζήτησης
          -> expanded_nodes: Μια λίστα με τους κόμβους που έχουν ήδη επεκταθεί
          Επιστρέφει: None
          Η συνάρτηση αυτή πρέπει να καλεστεί τουλάχιστον μια φορά για να μπορέσει να σχεδιαστει ένα animation (πρεπεί το animation να έχει τουλάχιστον ένα frame).
        '''
        image = np.zeros((grid.N, grid.N, 3), dtype=int)
        image[grid.grid == 0] = [255, 255, 255]
        image[grid.grid == 1] = [0, 0, 0]
        
        for node in expanded_nodes:
            image[node] = [0, 0, 128]

        for node in frontier:
            image[node] = [0, 225, 0]

        image[self.S] = [50, 168, 64]
        image[self.F] = [168, 50, 50]
        self.images.append(image)
    
    def add_path(self, path):
        '''
          Η συνάρτηση αυτή προσθέτει στο τελευταίο frame το βέλτιστο μονοπάτι.
          Είσοδος:
          -> path: Μια λίστα η όποια περιέχει το βέλτιστο μονοπάτι (η οποία πρέπει να περιέχει και τον κόμβο αρχή και τον κόμβο στόχο)
          Έξοδος: None
        '''
        for n in path[1:-1]:
            image = np.copy(self.images[-1])
            image[n] = [66, 221, 245]
            self.images.append(image)
        for _ in range (100):
            self.images.append(image)
        
    def create_gif(self, fps = 30, repeat_delay = 2000):
        if len(self.images) == 0:
            raise EmptyStackOfImages("Error! You have to call 'draw_step' at  first.")
        fig = plt.figure()
        plt.axis('off')
        ims = []
        for img in self.images:
            img = plt.imshow(img)
            ims.append([img])
        ani = animation.ArtistAnimation(fig, ims, interval=1000//fps, blit=True, repeat_delay= repeat_delay)
        plt.close(fig)
        return ani
        
    def save_gif(self, filename, fps = 30):
        '''
            Η συνάρτηση αυτή ξαναδημιουργεί και αποθηκεύει το animation σε ένα αρχείο.
            Είσοδος:
            -> Το όνομα του αρχείου με κατάληξη .gif
            Έξοδος: (None)
        '''
        ani = self.create_gif(fps)
        writer = PillowWriter(fps= fps)
        ani.save(filename, writer=writer)

    def show_gif(self, fps= 30, repeat_delay = 2000):
        '''
            Η συνάρτηση αυτή εμφανίζει inline το animation.
            Είσοδος:
            -> fps: τα frames per second
            Έξοδος: Το αντικείμενο που παίζει το animation
            Exceptions: EmptyStackOfImages αν το animation δεν έχει ούτε ένα frame, δηλαδή αν η draw_step δεν έχει καλεστεί ποτέ.
        '''
        ani = self.create_gif(fps, repeat_delay)
        # return HTML(ani.to_html5_video())
        return HTML(ani.to_jshtml())

    def show_last_frame(self):
        '''
            Η μέθοδος αυτή εμφανίζει inline το τελευταίο frame που έχει δημιουργήθει.
            Είσοδος:
            Έξοδος: Το αντικείμενο που εμφανίζει την εικόνα.
            Exceptions: EmptyStackOfImages αν το animation δεν έχει ούτε ένα frame, δηλαδή αν η draw_step δεν έχει καλεστεί ποτέ.
        '''
        if len(self.images) == 0:
            raise EmptyStackOfImages("Error! You have to call 'draw_step' at  first.")
        else:
            plt.imshow(self.images[-1])


class EmptyStackOfImages(Exception):
    pass
````

````
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import random

class grid:
    def __init__(self, N, S, F):
        
        ## Make sure start and end are within the grid
        assert N > 2
        assert S[0] < N
        assert S[1] < N
        assert F[0] < N
        assert F[1] < N

        assert S[0] > 0
        assert S[1] > 0
        assert F[0] > 0
        assert F[1] > 0

        self.N = N
        ## Initialize grid with obstacles
        self.grid = np.ones((N, N), dtype=np.int32)

        ### solution
        frontiers = set()
        neighbours = set()

        # adds to set frontiers all the cells in distance 2 from S provided they are part of the wall
        def addfrontiers(S):
          if S[1]-2 >= 0:
            if self.grid[S[0], S[1]-2] == 1:
              frontiers.add((S[0], S[1]-2))
          if S[1]+2 < N:
            if(self.grid[S[0], S[1]+2] == 1):
              frontiers.add((S[0], S[1]+2))
          if S[0]-2 >= 0:
            if self.grid[S[0]-2, S[1]] == 1:
              frontiers.add((S[0]-2, S[1]))
          if S[0]+2 < N:
            if self.grid[S[0]+2, S[1]] == 1:
              frontiers.add((S[0]+2, S[1]))

        # adds to set neighbours all the cells in distance 2 from S provided they are part of the maze
        def addneighbours(S):
          if S[1]-2 >= 0:
            if self.grid[S[0], S[1]-2] == 0:
              neighbours.add((S[0], S[1]-2))
          if S[1]+2 < N:
            if self.grid[S[0], S[1]+2] == 0:
              neighbours.add((S[0], S[1]+2))
          if S[0]-2 >= 0:
            if self.grid[S[0]-2, S[1]] == 0:
              neighbours.add((S[0]-2, S[1]))
          if S[0]+2 < N:
            if self.grid[S[0]+2, S[1]] == 0:
              neighbours.add((S[0]+2, S[1]))

        # makes sure that S,F aren't trapped somewhere.
        def checkCross(S):
          border = set()
          borderCnt = 0
          
          adjCells = [(S[0]-1,S[1]), (S[0]+1,S[1]), (S[0],S[1]-1), (S[0],S[1]+1)]
          for i in adjCells:
            if i[0] >= 0 and i[0] < N and i[1] >= 0 and i[1] < N:
              if self.grid[i] == 1:
                border.add(i)
                borderCnt += 1
            else:
              borderCnt += 1

          if borderCnt == 4:
            randBorder = random.choice(tuple(border))
            self.grid[randBorder[0], randBorder[1]] = 0

        #starting the algorithm from S ensures that it won't be blocked
        frontiers.add((random.randrange(0,N), random.randrange(0,N)))

        # while set frontier contains elements, randomly choose a frontier and turn it into a path cell.
        # In order to add to it to the rest of the maze find a neighbouring cell (path cells with distance 2) that is a part of the maze 
        # and connect them by turning the wall cell that separates them into a path cell.
        # Then add the frontiers (wall cells with distance 2) of the current frontier to the set of frontiers.
        while frontiers:
          randFrontier = random.choice(tuple(frontiers))
          self.grid[randFrontier] = 0
          neighbours.clear()
          addneighbours(randFrontier)
          if neighbours:
            randNeighbour = random.choice(tuple(neighbours))
            self.grid[((randFrontier[0] + randNeighbour[0])//2), ((randFrontier[1] + randNeighbour[1])//2)] = 0
          addfrontiers(randFrontier)
          frontiers.remove(randFrontier)

        self.grid[S] = 0
        self.grid[F] = 0
        checkCross(S)
        checkCross(F)
        
        # create more paths, by randomly deleting obstacles.        
        for i in range (N**2//10):
          randx = random.randrange(0,N)
          randy = random.randrange(0,N)
          self.grid[randx][randy] = 0

    def draw_map(self, S=None, F=None, path=None):
        image = np.zeros((self.N, self.N, 3), dtype=int)
        image[self.grid == 0] = [255, 255, 255]
        image[self.grid == 1] = [0, 0, 0]
        if S:
            image[S] = [50, 168, 64]
        if F:
            image[F] = [168, 50, 50]
        if path:
            for n in path[1:-1]:
                image[n] = [66, 221, 245]

        plt.imshow(image)
        plt.xticks([])
        plt.yticks([])
        plt.show()      
````

````
class Node:
  def __init__(self, position, parent):
        self.position = position
        self.parent = parent
        self.g = 0
        self.h = 0
        self.f = 0
  
  def __eq__(self,other):
    return self.position == other.position

class pathfinder:
    def __init__(self, S, F, grid, c, h):
        self.S = S
        self.F = F
        self.grid = grid
        self.cost = c
        self.heuristic = h
        #for visualization 
        self.vis = visualization(S, F)
        self.path = []
 
    def find_path(self):
        ## YOUR CODE HERE
        openList = []
        closedList = []

        start = Node(S, None)
        goal = Node(F, None)

        openList.append(start)
        
        while openList:
          temp = openList[0]
          idx = 0
          for i in range (len(openList)):
            if openList[i].f < temp.f:
              idx = i

          current = openList.pop(idx)
          closedList.append(current)

          if current == goal:
            p = []
            while current != start:
              p.append(current.position)
              current = current.parent
            p.append(current.position)
            self.path = p[::-1]
            return

          (x, y) = current.position
          for adjacent in [(x-1,y), (x+1,y), (x,y-1), (x,y+1)]:
            
            (adjX, adjY) = adjacent

            if adjX < 0 or adjX >= N or adjY < 0 or adjY >= N:
              continue

            if self.grid.grid[adjX][adjY] == 1:
              continue

            adjNode = Node(adjacent, current)
            if adjNode in closedList:
              continue
            
            adjNode.g = self.cost(adjX, adjY) + current.g
            adjNode.h = self.heuristic(adjX, adjY)
            adjNode.f = adjNode.g + adjNode.h
            
            tempG = 0
            tempG_idx = 0
            for i in range(len(openList)):
              if adjNode == openList[i]:
                tempG = openList[i].g
                tempG_idx = i
                continue

            if tempG > 0 and adjNode.g <= tempG:
              openList.pop(tempG_idx)
              openList.append(adjNode)
            elif tempG == 0:
              openList.append(adjNode)
            continue

    def get_path(self):
        self.find_path()
        return self.path
````

````
import math

N = 100
S = (5, 9)
F = (95, 85)

map = grid(N, S, F)
# g(n) = 0 and h(n) = manhattan
pf1 = pathfinder(S, F, map, lambda x, y: 0, lambda x, y: abs(x-F[0]) + abs(y-F[1]))
print("g(n) = 0 and h(n) = manhattan")
map.draw_map(S, F, pf1.get_path())
# g(n) = 0 and h(n) = euclidean
pf2 = pathfinder(S, F, map, lambda x, y: 0, lambda x, y: math.sqrt((x-F[0])**2 + (y-F[1])**2))
print("g(n) = 0 and h(n) = euclidean")
map.draw_map(S, F, pf2.get_path())
# g(n) = 1 and h(n) = 0
pf3 = pathfinder(S, F, map, lambda x, y: 1, lambda x, y: 0)
print("g(n) = 1 and h(n) = 0")
map.draw_map(S, F, pf3.get_path())
# g(n) = 1 and h(n) = manhattan
pf4 = pathfinder(S, F, map, lambda x, y: 1, lambda x, y: abs(x-F[0]) + abs(y-F[1]))
print("g(n) = 1 and h(n) = manhattan")
map.draw_map(S, F, pf4.get_path())
# g(n) = 1 and h(n) = euclidean
pf5 = pathfinder(S, F, map, lambda x, y: 1, lambda x, y: math.sqrt((x-F[0])**2 + (y-F[1])**2))
print("g(n) = 1 and h(n) = euclidean")
map.draw_map(S, F, pf5.get_path())
````

````
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import random

class cell:
    def __init__(self,trans_cost):
      self.obstacle = 1
      self.trans_cost = trans_cost

class grid:
    def __init__(self, N, S, F, cost_fun):
        
        ## Make sure start and end are within the grid
        assert N > 2
        assert S[0] < N
        assert S[1] < N
        assert F[0] < N
        assert F[1] < N

        assert S[0] > 0
        assert S[1] > 0
        assert F[0] > 0
        assert F[1] > 0

        self.N = N
        self.cost_fun = cost_fun
        ## Initialize grid with obstacles
        #self.grid = np.ones((N, N), dtype=np.int32)

        self.grid = [[cell(self.cost_fun()) for j in range(N)] for i in range(N)]

        ### solution
        frontiers = set()
        neighbours = set()

        # adds to set frontiers all the cells in distance 2 from S provided they are part of the wall
        def addfrontiers(S):
          if S[1]-2 >= 0:
            if self.grid[S[0]][S[1]-2].obstacle == 1:
              frontiers.add((S[0], S[1]-2))
          if S[1]+2 < N:
            if self.grid[S[0]][S[1]+2].obstacle == 1:
              frontiers.add((S[0], S[1]+2))
          if S[0]-2 >= 0:
            if self.grid[S[0]-2][S[1]].obstacle == 1:
              frontiers.add((S[0]-2, S[1]))
          if S[0]+2 < N:
            if self.grid[S[0]+2][S[1]].obstacle == 1:
              frontiers.add((S[0]+2, S[1]))

        # adds to set neighbours all the cells in distance 2 from S provided they are part of the maze
        def addneighbours(S):
          if S[1]-2 >= 0:
            if self.grid[S[0]][S[1]-2].obstacle == 0:
              neighbours.add((S[0], S[1]-2))
          if S[1]+2 < N:
            if self.grid[S[0]][S[1]+2].obstacle == 0:
              neighbours.add((S[0], S[1]+2))
          if S[0]-2 >= 0:
            if self.grid[S[0]-2][S[1]].obstacle == 0:
              neighbours.add((S[0]-2, S[1]))
          if S[0]+2 < N:
            if self.grid[S[0]+2][S[1]].obstacle == 0:
              neighbours.add((S[0]+2, S[1]))

        # makes sure that S,F aren't trapped somewhere.
        def checkCross(S):
          border = set()
          borderCnt=0
          
          adjCells=[(S[0]-1,S[1]), (S[0]+1,S[1]), (S[0],S[1]-1), (S[0],S[1]+1)]
          for i in adjCells:
            if i[0] >= 0 and i[0] < N and i[1] >= 0 and i[1] < N:
              if self.grid[i[0]][i[1]].obstacle == 1:
                border.add(i)
                borderCnt += 1
            else:
              borderCnt += 1

          if borderCnt == 4:
            randBorder = random.choice(tuple(border))
            self.grid[randBorder[0]][randBorder[1]].obstacle = 0

        #starting the algorithm from S ensures that it won't be blocked
        frontiers.add((random.randrange(0,N), random.randrange(0,N)))

        # while set frontier contains elements, randomly choose a frontier and turn it into a path cell.
        # In order to add to it to the rest of the maze find a neighbouring cell (path cells with distance 2) that is a part of the maze 
        # and connect them by turning the wall cell that separates them into a path cell.
        # Then add the frontiers (wall cells with distance 2) of the current frontier to the set of frontiers.
        while frontiers:
          randFrontier = random.choice(tuple(frontiers))
          self.grid[randFrontier[0]][randFrontier[1]].obstacle = 0
          neighbours.clear()
          addneighbours(randFrontier)
          if neighbours:
            randNeighbour = random.choice(tuple(neighbours))
            self.grid[((randFrontier[0] + randNeighbour[0])//2)][((randFrontier[1] + randNeighbour[1])//2)].obstacle = 0
          addfrontiers(randFrontier)
          frontiers.remove(randFrontier)

        self.grid[S[0]][S[1]].obstacle = 0
        self.grid[F[0]][F[1]].obstacle = 0
        checkCross(S)
        checkCross(F)

        # create more paths, by randomly deleting obstacles.        
        for i in range (N**2//10):
          randx = random.randrange(0,N)
          randy = random.randrange(0,N)
          self.grid[randx][randy].obstacle = 0

    def draw_map(self, S=None, F=None, path=None):
        image = np.zeros((self.N, self.N, 3), dtype=int)
        for i in range (self.N):
          for j in range (self.N):
            if self.grid[i][j].obstacle == 0:
              image[i,j] = [255, 255, 255]
            else:
              image[i,j] = [0, 0, 0]
        if S:
            image[S] = [50, 168, 64]
        if F:
            image[F] = [168, 50, 50]
        if path:
            for n in path[1:-1]:
                image[n] = [66, 221, 245]

        plt.imshow(image)
        plt.xticks([])
        plt.yticks([])
        plt.show()      
````

````
# from visualization import *
class Node:
  def __init__(self, position, parent):
        self.position = position
        self.parent = parent
        self.g = 0
        self.h = 0
        self.f = 0
  
  def __eq__(self,other):
    return self.position == other.position

class pathfinder:
    def __init__(self, S, F, grid, h):
        self.S = S
        self.F = F
        self.grid = grid
        #self.cost = c
        self.heuristic = h
        #for visualization 
        self.vis = visualization(S, F)
        self.path=[]
        self.closedListLength = 0
        
 
    def find_path(self):
        ## YOUR CODE HERE
        openList = []
        closedList = []

        start = Node(S, None)
        goal = Node(F, None)

        openList.append(start)
        
        while openList:
          temp = openList[0]
          idx = 0
          for i in range (len(openList)):
            if openList[i].f < temp.f:
              idx = i

          current = openList.pop(idx)
          closedList.append(current)

          if current == goal:
            self.closedListLength=len(closedList)
            p = []
            while current != start:
              p.append(current.position)
              current = current.parent
            p.append(current.position)
            self.path = p[::-1]
            return

          (x, y) = current.position
          for adjacent in [(x-1,y), (x+1,y), (x,y-1), (x,y+1)]:
            
            (adjX, adjY) = adjacent

            if adjX < 0 or adjX >= N or adjY < 0 or adjY >= N:
              continue

            if self.grid.grid[adjX][adjY].obstacle == 1:
              continue

            adjNode = Node(adjacent, current)
            if adjNode in closedList:
              continue
            
            #adjNode.g = self.cost(adjX, adjY) + current.g
            adjNode.g = self.grid.grid[adjX][adjY].trans_cost + current.g
            adjNode.h = self.heuristic(adjX, adjY)
            adjNode.f = adjNode.g + adjNode.h
            
            tempG = 0
            tempG_idx = 0
            for i in range(len(openList)):
              if adjNode==openList[i]:
                tempG = openList[i].g
                tempG_idx=i
                continue

            if tempG > 0 and adjNode.g <= tempG:
              openList.pop(tempG_idx)
              openList.append(adjNode)
            elif tempG == 0:
              openList.append(adjNode)
            continue

    def get_path(self):
        self.find_path()
        return (len(self.path),self.closedListLength)
````

````
import math
 
maxGridSize = 50

path_size = np.zeros((9,maxGridSize//5), dtype=np.int32)
clList_size = np.zeros((9,maxGridSize//5), dtype=np.int32)
i=0

for cnt in range (5,maxGridSize+1,5):
  N = cnt
  S = (2, 2)
  F = (cnt-2, cnt-2)
  
  # g(n) = uniform
  map = grid(N, S, F, lambda: np.random.uniform())
  pf1 = pathfinder(S, F, map, lambda x, y: abs(x-F[0]) + abs(y-F[1]))
  (path_size[0][i], clList_size[0][i]) = pf1.get_path()
  pf2 = pathfinder(S, F, map, lambda x, y: math.sqrt((x-F[0])**2 + (y-F[1])**2))
  (path_size[1][i], clList_size[1][i]) = pf2.get_path()
  pf3 = pathfinder(S, F, map, lambda x, y: 0)
  (path_size[2][i], clList_size[2][i]) = pf3.get_path()

  # g(n) = beta
  map2 = grid(N, S, F, lambda: np.random.beta(2,2))
  pf4 = pathfinder(S, F, map2, lambda x, y: abs(x-F[0]) + abs(y-F[1]))
  (path_size[3][i], clList_size[3][i]) = pf4.get_path()
  pf5 = pathfinder(S, F, map2, lambda x, y: math.sqrt((x-F[0])**2 + (y-F[1])**2))
  (path_size[4][i], clList_size[4][i]) = pf5.get_path()
  pf6 = pathfinder(S, F, map2, lambda x, y: 0)
  (path_size[5][i], clList_size[5][i]) = pf6.get_path()

  # g(n) = triangular
  map3 = grid(N, S, F, lambda: np.random.triangular(0,0.5,1))
  pf7 = pathfinder(S, F, map3, lambda x, y: abs(x-F[0]) + abs(y-F[1]))
  (path_size[6][i], clList_size[6][i]) = pf7.get_path()
  pf8 = pathfinder(S, F, map3, lambda x, y: math.sqrt((x-F[0])**2 + (y-F[1])**2))
  (path_size[7][i], clList_size[7][i]) = pf8.get_path()
  pf9 = pathfinder(S, F, map3, lambda x, y: 0)
  (path_size[8][i], clList_size[8][i]) = pf9.get_path()
  
  i+=1

xaxis = np.arange(0.0, 50.0, 5)
legend_list = ["g(n) = uniform, h(n) = manhattan","g(n) = uniform, h(n) = euclidean","g(n) = uniform, h(n) = 0",
             "g(n) = beta, h(n) = manhattan","g(n) = beta, h(n) = euclidean","g(n) = beta, h(n) = 0",
             "g(n) = triangular, h(n) = manhattan","g(n) = triangular, h(n) = euclidean","g(n) = triangular, h(n) = 0"]
plt.figure(figsize = (15,8))
plt.xlabel("Grid size")
plt.ylabel("Path size")
for i in range (0,9):
  plt.plot(xaxis, path_size[i], label = legend_list[i])
plt.legend()

plt.figure(figsize = (15,8))
plt.xlabel("Grid size")
plt.ylabel("Number of expanded nodes")
for i in range (0,9):
  plt.plot(xaxis, clList_size[i], label = legend_list[i])
plt.legend()
````
