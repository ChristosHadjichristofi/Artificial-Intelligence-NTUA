### Αλγόριθμος παραγωγής λαβυρίνθων
```
1. Αρχικά όλα τα κελιά περιέχουν εμπόδια και η λίστα των τοίχων είναι κενή.
2. Ένα κελί επιλέγεται ως μέρος του λαβυρίνθου. Οι τοίχοι του κελιού προστίθενται στη λίστα των τοίχων.
3. Όσο η λίστα των τοίχων δεν είναι κενή:
      1. Επιλέγεται τυχαία ένας τοίχος από τη λίστα. Εαν μόνο ένα από τα (δύο) κελιά που χωρίζει ο τοίχος είναι μέρος του λαβυρίνθου, τότε:
            1. Ο τοίχος και το κελί που δεν είναι μέρος του λαβυρίνθου, γίνονται μέρος του λαβυρίνθου.
            2. Οι τοίχοι των νέων κελιών που προστέθηκαν στο λαβύρινθο προστίθενται στη λίστα των τοίχων.
      2. Ο τοίχος αφαιρείται από τη λίστα των τοίχων
```
*ως τοίχοι ενός κελιού θεωρούνται όσα απο τα 4 γειτονικά κελιά (όχι τα διαγώνια) περιέχουν εμπόδιο

# Κώδικας visualization


````
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter
from IPython.display import HTML

class visualization:
    def __init__(self, S, F):
        '''
          Η μέθοδος αυτή αρχικοποιεί ένα αντικείμενο τύπου visualization.
          Είσοδος: 
          -> S: το σημείο εκκίνσης της αναζήτησης
          -> F: το σημείο τερματισμού
        '''
        self.S = S
        self.F = F
        self.images = []
    
    def draw_step(self, grid, frontier, expanded_nodes):
        '''
          Η συνάρτηση αυτή καλείται για να σχεδιαστεί ένα frame στο animation (πρακτικά έπειτα από την επέκταση κάθε κόμβου)
          Είσοδος: 
          -> grid: Ένα χάρτης τύπου grid
          -> frontier: Μια λίστα με τους κόμβους που ανήκουν στο μέτωπο της αναζήτησης
          -> expanded_nodes: Μια λίστα με τους κόμβους που έχουν ήδη επεκταθεί
          Επιστρέφει: None
          Η συνάρτηση αυτή πρέπει να καλεστεί τουλάχιστον μια φορά για να μπορέσει να σχεδιαστει ένα animation (πρεπεί το animation να έχει τουλάχιστον ένα frame).
        '''
        image = np.zeros((grid.N, grid.N, 3), dtype=int)
        
        for i in range (grid.N):
          for j in range (grid.N):
            if grid.grid.grid[i][j].obstacle == 0:
              image[i,j] = [255, 255, 255]
            else:
              image[i,j] = [0, 0, 0]
        
        for node in expanded_nodes:
            image[node] = [0, 0, 128]

        for node in frontier:
            image[node] = [0, 225, 0]

        image[self.S] = [50, 168, 64]
        image[self.F] = [168, 50, 50]
        self.images.append(image)

    def add_path(self, path):
        '''
          Η συνάρτηση αυτή προσθέτει στο τελευταίο frame το βέλτιστο μονοπάτι.
          Είσοδος:
          -> path: Μια λίστα η όποια περιέχει το βέλτιστο μονοπάτι (η οποία πρέπει να περιέχει και τον κόμβο αρχή και τον κόμβο στόχο)
          Έξοδος: None
        '''
        for n in path[1:-1]:
            image = np.copy(self.images[-1])
            image[n] = [66, 221, 245]
            self.images.append(image)
        for _ in range (100):
            self.images.append(image)
        
    def create_gif(self, fps = 30, repeat_delay = 2000):
        if len(self.images) == 0:
            raise EmptyStackOfImages("Error! You have to call 'draw_step' at  first.")
        fig = plt.figure()
        plt.axis('off')
        ims = []
        for img in self.images:
            img = plt.imshow(img)
            ims.append([img])
        ani = animation.ArtistAnimation(fig, ims, interval=1000//fps, blit=True, repeat_delay= repeat_delay)
        plt.close(fig)
        return ani
        
    def save_gif(self, filename, fps = 30):
        '''
            Η συνάρτηση αυτή ξαναδημιουργεί και αποθηκεύει το animation σε ένα αρχείο.
            Είσοδος:
            -> Το όνομα του αρχείου με κατάληξη .gif
            Έξοδος: (None)
        '''
        ani = self.create_gif(fps)
        writer = PillowWriter(fps= fps)
        ani.save(filename, writer=writer)

    def show_gif(self, fps= 30, repeat_delay = 2000):
        '''
            Η συνάρτηση αυτή εμφανίζει inline το animation.
            Είσοδος:
            -> fps: τα frames per second
            Έξοδος: Το αντικείμενο που παίζει το animation
            Exceptions: EmptyStackOfImages αν το animation δεν έχει ούτε ένα frame, δηλαδή αν η draw_step δεν έχει καλεστεί ποτέ.
        '''
        ani = self.create_gif(fps, repeat_delay)
        # return HTML(ani.to_html5_video())
        return HTML(ani.to_jshtml())

    def show_last_frame(self):
        '''
            Η μέθοδος αυτή εμφανίζει inline το τελευταίο frame που έχει δημιουργήθει.
            Είσοδος:
            Έξοδος: Το αντικείμενο που εμφανίζει την εικόνα.
            Exceptions: EmptyStackOfImages αν το animation δεν έχει ούτε ένα frame, δηλαδή αν η draw_step δεν έχει καλεστεί ποτέ.
        '''
        if len(self.images) == 0:
            raise EmptyStackOfImages("Error! You have to call 'draw_step' at  first.")
        else:
            plt.imshow(self.images[-1])


class EmptyStackOfImages(Exception):
    pass
````

````
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import random
import math

class cell:
    def __init__(self,trans_cost):
      self.obstacle = 1
      self.trans_cost = trans_cost

class grid:
    def __init__(self, N, S, F, cost_fun):
        
        ## Make sure start and end are within the grid
        assert N > 2
        assert S[0] < N
        assert S[1] < N
        assert F[0] < N
        assert F[1] < N

        assert S[0] > 0
        assert S[1] > 0
        assert F[0] > 0
        assert F[1] > 0

        self.N = N
        self.cost_fun = cost_fun
        ## Initialize grid with obstacles
        #self.grid = np.ones((N, N), dtype=np.int32)
        self.grid = [[cell(self.cost_fun()) for j in range(N)] for i in range(N)]

        frontiers = set()
        neighbours = set()

        # adds to set frontiers all the cells in distance 2 from S provided they are part of the wall
        def addfrontiers(S):
          for (x,y) in [(S[0]-2,S[1]), (S[0]+2,S[1]), (S[0],S[1]-2), (S[0],S[1]+2)]:
            if x >= 0 and x < N and y >= 0 and y < N:
              if self.grid[x][y].obstacle == 1:
                frontiers.add((x,y))

        # adds to set neighbours all the cells in distance 2 from S provided they are part of the maze
        def addneighbours(S): 
          for (x,y) in [(S[0]-2,S[1]), (S[0]+2,S[1]), (S[0],S[1]-2), (S[0],S[1]+2)]:
            if x >= 0 and x < N and y >= 0 and y < N:
              if self.grid[x][y].obstacle == 0:
                neighbours.add((x,y))

        # makes sure that S,F aren't trapped somewhere.
        def checkCross(S):
          border = set()
          borderCnt = 0
          
          adjCells = [(S[0]-1,S[1]), (S[0]+1,S[1]), (S[0],S[1]-1), (S[0],S[1]+1)]
          for i in adjCells:
            if i[0] >= 0 and i[0] < N and i[1] >= 0 and i[1] < N:
              if self.grid[i[0]][i[1]].obstacle == 1:
                border.add(i)
                borderCnt += 1
            else:
              borderCnt += 1

          if borderCnt == 4:
            randBorder = random.choice(tuple(border))
            self.grid[randBorder[0]][randBorder[1]].obstacle = 0

        # Start creating the maze from a random cell
        frontiers.add((random.randrange(0,N), random.randrange(0,N)))

        # while set frontier contains elements, randomly choose a frontier and turn it into a path cell.
        # In order to add to it to the rest of the maze find a neighbouring cell (path cells with distance 2) that is a part of the maze 
        # and connect them by turning the wall cell that separates them into a path cell.
        # Then add the frontiers (wall cells with distance 2) of the current frontier to the set of frontiers.
        while frontiers:
          randFrontier = random.choice(tuple(frontiers))
          self.grid[randFrontier[0]][randFrontier[1]].obstacle = 0
          neighbours.clear()
          addneighbours(randFrontier)
          if neighbours:
            randNeighbour = random.choice(tuple(neighbours))
            self.grid[((randFrontier[0] + randNeighbour[0])//2)][((randFrontier[1] + randNeighbour[1])//2)].obstacle = 0
          addfrontiers(randFrontier)
          frontiers.remove(randFrontier)

        self.grid[S[0]][S[1]].obstacle = 0
        self.grid[F[0]][F[1]].obstacle = 0
        checkCross(S)
        checkCross(F)

        # create more paths, by randomly deleting obstacles.        
        for i in range (N**2//10):
          randx = random.randrange(0,N)
          randy = random.randrange(0,N)
          self.grid[randx][randy].obstacle = 0

    # draw_map will also draw the expanded nodes and the frontier nodes
    def draw_map(self, S = None, F = None, path = None, frontier = [], expanded_nodes = []):
        image = np.zeros((self.N, self.N, 3), dtype=int)
        for i in range (self.N):
          for j in range (self.N):
            if self.grid[i][j].obstacle == 0:
              image[i,j] = [255, 255, 255]
            else:
              image[i,j] = [0, 0, 0]

        for node in expanded_nodes:
            image[node] = [0, 0, 128]

        for node in frontier:
            image[node] = [0, 225, 0]

        if S:
            image[S] = [50, 168, 64]
        if F:
            image[F] = [168, 50, 50]
        if path:
            for n in path[1:-1]:
                image[n] = [66, 221, 245]

        plt.imshow(image)
        plt.xticks([])
        plt.yticks([])
        plt.show()      
````

## Μέρος 2


````
# each cell is represented as a node with the following attributes
class Node:
  def __init__(self, position, parent):
        self.position = position
        self.parent = parent
        self.g = 0
        self.h = 0
        self.f = 0

  # overriding function equal to return true if two nodes have same position
  def __eq__(self,other):
    return self.position == other.position

class pathfinder:
    # added N, closedListLength, frontier, expanded_nodes part and gif attributes.
    # part -> when given "part2" it uses cost function given to calculate cost for nodes
    #         when given "part3" it uses the costs that are assigned from part1
    # gif -> determines whether it creates a gif or not
    def __init__(self, N, S, F, grid, c, h, part, gif):
        self.N = N
        self.S = S
        self.F = F
        self.grid = grid
        self.cost = c
        self.heuristic = h
        self.path = []
        self.closedListLength = 0
        self.frontier = []
        self.expanded_nodes = []
        self.part = part
        self.gif = gif        
        #for visualization 
        self.vis = visualization(S, F)
        
 
    def find_path(self):
        openList = []
        closedList = []
        self.frontier = []
        self.expanded_nodes = []

        # Create start and goal Nodes
        start = Node(S, None)
        goal = Node(F, None)

        # Append start node to the open list
        openList.append(start)
        self.frontier.append(S)
        
        # while openlist is not empty
        while openList:

          # find the node with the smallest f = g+h
          temp = openList[0]
          idx = 0
          for i in range (len(openList)):
            if openList[i].f < temp.f:
              idx = i

          # pop that node, set it as current and append it to the closed list
          current = openList.pop(idx)
          self.frontier.remove(current.position)
          closedList.append(current)
          self.expanded_nodes.append(current.position)

          if self.gif == "yes":
            self.vis.draw_step(self, self.frontier, self.expanded_nodes)

          # if this specific node is the goal, then find the path and return it
          if current == goal:
            if self.gif == "no":
              self.vis.draw_step(self, self.frontier, self.expanded_nodes)
            self.closedListLength = len(closedList)
            p = []
            while current != start:
              p.append(current.position)
              current = current.parent
            p.append(current.position)
            self.path = p[::-1]
            return

          # get x,y coordinates of current node
          (x, y) = current.position
          # for every adjacent node of this current
          for adjacent in [(x-1,y), (x+1,y), (x,y-1), (x,y+1)]:
            
            (adjX, adjY) = adjacent
            # check if adjacnet node is within bounds 
            if adjX < 0 or adjX >= N or adjY < 0 or adjY >= N:
              continue
            
            # if is obstacle, continue with the next adjacent node
            if self.grid.grid[adjX][adjY].obstacle == 1:
              continue

            # else if is within bounds and not obstacle, create adjacent Node
            adjNode = Node(adjacent, current)
            # if already traversed continue
            if adjNode in closedList:
              continue
            
            # if not in closed list create g,h,f of adjacent node
            # check which part of code will be used (using self.part) in order to use the right g function
            if self.part == "part2":
              adjNode.g = self.cost(adjX, adjY) + current.g
            else:
              adjNode.g = self.grid.grid[adjX][adjY].trans_cost + current.g
            adjNode.h = self.heuristic(adjX, adjY)
            adjNode.f = adjNode.g + adjNode.h
            
            # find if adjacent node is already in open list
            tempG = 0
            tempG_idx = 0
            tempG_pos = (0,0)
            for i in range(len(openList)):
              if adjNode == openList[i]:
                tempG = openList[i].g
                tempG_idx = i
                tempG_pos = openList[i].position
                continue

            # if adjacent node is already in open list and adjacent node's 
            # g is smaller than the one in the open list, replace it
            if tempG > 0 and adjNode.g <= tempG:
              self.frontier.remove(tempG_pos)
              openList.pop(tempG_idx)
              openList.append(adjNode)
              self.frontier.append(adjNode.position)
            # if not found in open list, add it to the open list
            elif tempG == 0:
              openList.append(adjNode)
              self.frontier.append(adjNode.position)
            continue

    def get_path(self):
        # call find_path() to run a star algorithm
        self.find_path()
        # for part 2 return the path, for part 3 return the length of the path and the closed list
        if self.part == "part2":
          return self.path
        else:
          return (len(self.path), self.closedListLength)
````

````
# After reading the example of visualization that was given to us
# we created an example of how one can create a gif in the modified code.
# gif is created and saved in the directory content\example.gif and its final frame is shown below

N = 16
S = (2, 2)
F = (14, 15)

manhattan = lambda x, y: abs(x-F[0]) + abs(y-F[1])
eucledian = lambda x, y: math.sqrt((x-F[0])**2 + (y-F[1])**2)

def showMapSolution(g_function, h_function):
  pf = pathfinder(N, S, F, map, g_function, h_function, "part2", "yes")
  pf.vis.add_path(pf.get_path())
  pf.vis.save_gif("example.gif")
  pf.vis.show_last_frame()

map = grid(N, S, F, lambda: 0)
print("g(n) = 0 and h(n) = manhattan")
showMapSolution(lambda x, y: 0, manhattan)
````

**g(n) = 0 και h(n) = manhattan ή euclidean:**

> Σε αυτή την περίπτωση ο αλγόριθμος που προκύπτει είναι ο **Best First** διότι κάθε φορά επεκτείνει τον κόμβο με τη μικρότερη τιμή της ευριστικής συνάρτηση και μόλις εντοπίσει τον τελικό κόμβο τερματίζει. Αυτός ο αλγόριθμος δεν εγγυάται τη βέλτιστη διαδρομή κάτι το οποίο φαίνεται και στην οπτικοποίηση της λύσης.

**g(n) = 1 και h(n) = 0:**

> Σε αυτή την περίπτωση ο αλγόριθμος που προκύπτει είναι ο **Dijkstra** διότι από έναν αρχικό κόμβο βρίσκει το βέλτιστο μονοπάτι προς όλους τους υπόλοιπους και σταματά όταν βρει τον τελικό κόμβο. Ο αλγόριθμος Dijkstra βρίσκει τη βέλτιστη διαδρομή όμως απαιτεί σημαντικά περισσότερο χρόνο από τις υπόλοιπες περιπτώσεις γιατί επεκτείνει πολύ περισσότερους κόμβους.

**g(n) = 1 και h(n) = manhattan ή euclidean:**

> Σε αυτή την περίπτωση ο αλγόριθμος που προκύπτει είναι ο **Α\*** διότι κάθε φορά επεκτείνει τον κόμβο με το μικρότερο άθροισμα τιμής ευριστικής συνάρτησης και πραγματικής απόστασης από την αρχική κατάσταση. Επίσης αν κατασκευάσει μονοπάτι σε κόμβο για τον οποίο υπήρχε ήδη μονοπάτι, κρατάει το μικρότερο σε κόστος και "κλαδεύει" το μεγαλύτερο. Ο αλγόριθμος Α* βρίσκει τη βέλτιστη διαδρομή όταν ο ευριστικός μηχανισμός είναι αποδεκτός (δηλαδή η τιμή h(n) για κάθε κόμβο είναι μικρότερη ή ίση της πραγματικής). Σε αυτή την περίπτωση και οι δύο ευριστικοί μηχανισμοί είναι αποδεκτοί. Παρατηρείται ακόμη ότι παρόλο που αμφότερες οι ευριστικές συναρτήσεις έχουν ίδιο μήκος μονοπατιού, η manhattan είναι σημαντικά γρηγορότερη γιατί επεκτείνει λιγότερους κόμβους.

**Πιθανές εναλλακτικές ευριστικές συναρτήσεις:**


* Θα μπορούσε να χρησιμοποιηθεί οποιαδήποτε νόρμα τάξης p για τον $R ^2$ <br> 
όπου  $||h(x,y)||_p=(\mid x-F[0]\mid^p + \mid y-F[1]\mid^p)^{1/p}$
>Συγκεκριμένα θα μπορούσε να χρησιμοποιηθεί η νόρμα άπειρο (infinity norm)
$||h(x,y)||_∞=max\{\mid x-F[0]\mid , \mid y-F[1]\mid\}$<br><br>
Η πιο πάνω ευριστική είναι αποδεκτή γιατί η τιμή που επιστρέφει για κάθε σημείο είναι μικρότερη ή ίση της πραγματικής 

  *Οι συναρτήσεις manhattan και eucledian αντιστοιχούν στις νόρμες με τάξη 1 και 2 αντίστοιχα* 

  Πηγή: Θ. Μ. Ρασσιάς, Μαθηματικά ΙΙ, Εκδόσεις Τσότρας, Αθήνα, 2014

<br>

* Μία άλλη πιθανή ευριστική συνάρτηση που μπορεί να χρησιμοποιηθεί είναι η διαγώνια απόσταση. Είναι παρεμφερής με την απόσταση Manhattan όμως προορίζεται για λαβυρίνθους στους οποίους είναι εφικτή η μετάβαση και στα 8 γειτονικά κελιά κάθε κελιού. Στη συγκεκριμένη περίπτωση θα μπορούσε να χρησιμοποιηθεί και είναι αποδεκτή διότι η τιμή που δίνει για κάθε κόμβο είναι μικρότερη ή ίση της πραγματικής.
$h_{diagonal}(x,y)=(\mid x-F[0]\mid+\mid y-F[1]\mid) - (2 - \sqrt{2})*min(\mid x-F[0]\mid,\mid y-F[1]\mid) $

  Πηγή: http://theory.stanford.edu/~amitp/GameProgramming/Heuristics.html


<br>

* Επίσης θα μπορούσε να χρησιμοποιηθεί μία ευριστική συνάρτηση που να λαμβάνει υπόψη την πιθανότητα ένα κελί να είναι εμπόδιο.<br>
Είναι γνωστό ότι η συνάρτηση Manhattan υπολογίζει το βέλτιστο μήκος μονοπατιού για λαβύρινθο χωρίς εμπόδια. Η πραγματική απόσταση είναι ελαφρώς μεγαλύτερη από αυτή και η διαφορά αυξάνεται όσο αυξάνεται το μέγεθος του λαβυρίνθου. Αυτό συμβαίνει γιατί παρόλο που υπάρχουν πολλά μονοπάτια με το βέλτιστο μήκος, είναι μεγάλη η πιθανότητα να συναντούν όλα τουλάχιστον ένα εμπόδιο στη διαδρομή. Θα μπορούσε λοιπόν να χρησιμοποιηθεί μια ευριστική συνάρτηση που να χρησιμοποιεί τη συνάρτηση Manhattan και να προσθέτει κάποιο διορθωτικό παράγοντα. Αυτός ο παράγοντας έχει υπολογιστεί στο **παράρτημα** με τη χρήση προσομοίωσης για λαβυρίνθους με Ν από 5 έως 100.<br>
$h_{ManhattanWithFactor}(n)=manhattan(n)+factor(N)$

  *Η πιο πάνω συνάρτηση δίνει μεν καλύτερη προσέγγιση της πραγματικής απόστασης από ότι οι προηγούμενες όμως είναι πιθανοτική και δεν είναι αποδεκτή άρα δεν δίνει πάντα το καλύτερο μονοπάτι.*

````
import math

# use uniform distribution, triangular distribution with parameters (0,0.5,1) and beta distribution with parameters (2,2)
# https://en.wikipedia.org/wiki/Continuous_uniform_distribution
uniform = lambda: np.random.uniform()
# https://en.wikipedia.org/wiki/Triangular_distribution
triangular = lambda: np.random.triangular(0,0.5,1)
# https://en.wikipedia.org/wiki/Beta_distribution
beta = lambda: np.random.beta(2,2)


def visualizeGraphs(title):
  for j in range (2):
    plt.figure(figsize = (15,8))
    plt.title(title)
    for i in range (0,9):
      plt.xlabel("Grid size")
      plt.ylabel(ylabels[j])
      plt.plot(xaxis, content[j][i], label = legend_list[i])
      plt.legend()

def fill(map, h_function, i, j):
  pf = pathfinder(N, S, F, map, lambda:0, h_function, "part3", "no")
  (path_size[i][j], clList_size[i][j]) = pf.get_path()

maxGridSize = 50
i = 0

path_size = np.zeros((9,maxGridSize//5 + 1), dtype = np.int32)
clList_size = np.zeros((9,maxGridSize//5 + 1), dtype = np.int32)

for cnt in range (5, maxGridSize + 6, 5):
  N = cnt
  S = (2, 2)
  F = (cnt-2, cnt-2)
  map1 = grid(N, S, F, uniform)
  map2 = grid(N, S, F, beta)
  map3 = grid(N, S, F, triangular)
  parameters = [(map1, manhattan),(map1, eucledian),(map1, lambda x,y: 0),
                (map2, manhattan),(map2, eucledian),(map2, lambda x,y: 0),
                (map3, manhattan),(map3, eucledian),(map3, lambda x,y: 0)]
  k = 0
  for (map, h_function) in parameters:
    fill(map, h_function, k, i)
    k += 1
  i += 1

xaxis = np.arange(0.0, maxGridSize+1, 5)
legend_list = ["g(n) = uniform, h(n) = manhattan","g(n) = uniform, h(n) = euclidean","g(n) = uniform, h(n) = 0",
               "g(n) = beta, h(n) = manhattan","g(n) = beta, h(n) = euclidean","g(n) = beta, h(n) = 0",
               "g(n) = triangular, h(n) = manhattan","g(n) = triangular, h(n) = euclidean","g(n) = triangular, h(n) = 0"]
ylabels = ["Path size", "Number of expanded nodes"]
content = [path_size, clList_size]

visualizeGraphs("Transition cost per cell within (0,1)")
````

**Συμπεράσματα:**

Παρατηρείται ότι όταν δεν χρησιμοποιείται ευριστική συνάρτηση ο αλγόριθμος είναι πολύ πιο αργός γιατί επεκτείνει πολύ περισσότερους κόμβους. Από την άλλη το μήκος του μονοπατιού που επιστρέφει είναι μικρότερο.
Σε αυτή την περίπτωση οι δύο ευριστικές δεν είναι αποδεκτές και δεν επιστρέφουν το βέλτιστο μονοπάτι λόγω της τυχαιότητας της πραγματικής απόστασης. Μεταξύ των δύο ευριστικών συναρτήσεων παρατηρείται ότι το μήκος μονοπατιού που δημιουργεί η eucledian είναι μικρότερο από αυτό της Manhattan άρα και πιο κοντά στη βέλτιστη λύση παρόλο που επεκτείνουν περίπου ίδιο αριθμό κόμβων.



````
# use the same distribution but add 1 to get values between 1,2 and divide by 10 to get values between 0,0.1
uniform1 = lambda: np.random.uniform() + 1
triangular1 = lambda: np.random.triangular(1,1.5,2)
beta1 = lambda: np.random.beta(2,2) + 1
uniform2 = lambda: np.random.uniform() / 10
triangular2 = lambda: np.random.triangular(0,0.05,0.1)
beta2 = lambda: np.random.beta(2,2) / 10

alternative_h = [(uniform1, triangular1, beta1), (uniform2, triangular2, beta2)]
h_names = ["Transition cost per cell  within (1,2)", "Transition cost per cell  within [0,0.1)"]
for index in range (2):
  i = 0
  path_size = np.zeros((9,maxGridSize//5 + 1), dtype = np.int32)
  clList_size = np.zeros((9,maxGridSize//5 + 1), dtype = np.int32)
  for cnt in range (5, maxGridSize + 6, 5):
    N = cnt
    S = (2, 2)
    F = (cnt-2, cnt - 2)
    map1 = grid(N, S, F, alternative_h[index][0])
    map2 = grid(N, S, F, alternative_h[index][1])
    map3 = grid(N, S, F, alternative_h[index][2])

    parameters = [(map1, manhattan),(map1, eucledian),(map1, lambda x,y: 0),
                  (map2, manhattan),(map2, eucledian),(map2, lambda x,y: 0),
                  (map3, manhattan),(map3, eucledian),(map3, lambda x,y: 0)]
    k = 0
    for (map, h_function) in parameters:
      fill(map, h_function, k, i)
      k += 1
    i += 1
  content = [path_size, clList_size]
  visualizeGraphs(h_names[index])
````

**Συμπεράσματα:**

Το εύρος τιμών που παίρνει το κόστος μετάβασης από ένα κόμβο σε άλλο επηρεάζει το βάρος της συνάρτησης g έναντι της h στον υπολογισμό της f. Όταν το εύρος τιμών είναι στο (1,2) η συνάρτηση g (πραγματική απόσταση) επηρεάζει περισσότερο τον αλγόριθμο από ότι η συνάρτηση h (ευριστική) ενώ όταν το εύρος τιμών είναι [0,0.1) επηρεάζει περισσότερο η ευριστική συνάρτηση. 

Γι΄ αυτό στην πρώτη περίπτωση παρατηρείται ότι όλοι οι αλγόριθμοι καθυστερούν εξίσου και επεκτείνουν πολύ μεγάλο αριθμό κόμβων.
Από την άλλη όταν η ευριστική επηρεάζει περισσότερο από τη συνάρτηση g ο αριθμός των κόμβων που επεκτείνει ο αλγόριθμος είναι και πάλι μεγαλύτερος από ότι στην πρώτη περίπτωση χωρίς να δίνει μικρότερα μονοπάτια.

````
array=[]

for j in range(5,101,5):
  N = j
  S = (1, 1)
  F = (N-2, N-2)

  cnt=0
  times=1000
  for i in range (times):
    map = grid(N, S, F, lambda: 0)
    pf = pathfinder(N, S, F, map, lambda x, y: 1, lambda x, y: abs(x-F[0])+abs(y-F[1]), "part2", "no")
    pathi=pf.get_path()
    factor=len(pathi)-(2*N-5)
    if factor>0:
      cnt+=factor
  array.append(cnt/times)


xaxis = np.arange(0.0, 5*len(array), 5)
plt.figure(figsize = (15,8))
plt.title("Avg extra steps (factor) from the ideal path size (Manhattan) per grid size")
plt.xlabel("Grid size")
plt.ylabel("Avg extra steps")
plt.plot(xaxis, array)
plt.show()
````
